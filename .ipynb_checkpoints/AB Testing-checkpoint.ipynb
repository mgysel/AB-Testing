## Clean

### control_df/treatment_df

#### Summary of Data Issues
- Quality
    - Remove column names spaces and make them all lowercase
    - Change datatypes for StoreID, Postal Code, and Region
- Tidiness
    - Remove Unnecessary Columns: Right_Name, Phone Number, Street Combined, Street 1, Street 2, Street 3, Country, Coordinates, Latitude, Longitude, Current Timezone Offset, Olson Timezone

```python
control_df_clean = control_df.copy()
treatment_df_clean = treatment_df.copy()
```

#### Remove Unnecessary Columns

##### Define
Remove all unnecessary columns from control_df and treatment_df

##### Code

```python
drop_columns_control = ['Right_Name', 'Phone Number', 'Street Combined', 'Street 1', 
                'Street 2', 'Street 3', 'Country', 'Coordinates', 'Latitude',
                'Longitude', 'Current Timezone Offset', 'Olson Timezone']
drop_columns_treatment = ['Phone Number', 'Street Combined', 'Street 1', 
                'Street 2', 'Street 3', 'Country', 'Coordinates', 'Latitude',
                'Longitude', 'Current Timezone Offset', 'Olson Timezone', 'Name']
control_df_clean.drop(columns=drop_columns_control, inplace=True)
treatment_df_clean.drop(columns=drop_columns_treatment, inplace=True)
```

##### Test

```python
control_df_clean.head()
```

```python
treatment_df_clean.head()
```

#### Column Names

##### Define
Remove all spaces, make lowercase, and reorder the control_df and treatment_df titles

##### Code

```python
# control_df
control_df_clean.rename(columns = lambda x: x.lower().replace(' ','_'), inplace=True)
control_df_clean.rename(columns = {'storeid': 'store_id', 
                                   'avgmonthsales': 'avg_month_sales'}, inplace=True)

# treatment_df
treatment_df_clean.rename(columns = lambda x: x.lower().replace(' ','_'), inplace=True)
treatment_df_clean.rename(columns = {'storeid': 'store_id', 
                                   'avgmonthsales': 'avg_month_sales', 
                                   'right_state': 'state'}, inplace=True)
```

##### Test

```python
control_df_clean.head()
```

```python
control_df_clean.head()
```

#### Column Order

##### Define
Reorder treatment_df columns to match control_df columns

##### Code

```python
treatment_df_clean.columns
```

```python
treatment_df_clean = treatment_df_clean[['store_id', 'sq_ft', 'avg_month_sales', 
                                         'city', 'state', 'postal_code', 'region', 'timezone']]
```

##### Test

```python
treatment_df_clean.head()
```

#### Change datatypes for store_id, state, postal_code, region, timezone

##### Define
Change store_id to categorical, state to categorical, postal_code to categorical, region to categorical, and timezone to categorical

##### Code

```python
# columns to change to categorical
columns = ['store_id', 'state', 'postal_code', 'region', 'timezone']

# change column datatypes in control_df_clean and treatment_df_clean
for column in columns:
    control_df_clean[column] = control_df_clean[column].astype('category')
    treatment_df_clean[column] = treatment_df_clean[column].astype('category')
```

##### Test

```python
control_df_clean.info()
```

```python
treatment_df_clean.info()
```

### transactions_df

#### Remove Unnecessary Columns

Remove Invoice Number, SKU, QTY, Size

```python
# Columns to remove
columns = ['Invoice Number', 'SKU', 'QTY', 'Size']

# Remove each columns
transactions_df_clean.drop(columns=columns, inplace=True)
```

#### Change Column Names

##### Define
Remove spaces and lowercase column names

##### Code

##### Test

#### Change Datatypes
